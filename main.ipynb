```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 200)
```

```python
data = pd.read_csv('flights.csv')
print(data.shape)
data.head()
```

### Upon inspection we found out that there are alot of columns with Nan values or have most of the cells blank.


### Let's move ahead and remove those useless columns.

```python
data = data.dropna(axis=1, how='all')
print(data.shape)
```

```python
# check null values
for i in range(0, len(data.columns)):
    print(data.columns[i], data[data.columns[i]].isnull().sum())
```

```python
# check null values
for i in range(0, len(data.columns)):
    print(data.columns[i], data[data.columns[i]].isnull().sum())
```

```python
# Check Unique values in each column
for i in range(0, len(data.columns)):
    print(data.columns[i], data[data.columns[i]].nunique())
```

## Total Flights wrt to date and Months

```python
print("Data Timeline Information")

print("-------------------------")
print("Year")
print(data['year'].value_counts())
print("-------------------------")
print("Month")
print(data['month'].value_counts())
print("-------------------------")

plt.figure(figsize=(7, 3))
sns.countplot(data['month'])
plt.title('Month Wise Flight Count')
plt.show()

# Countplot day wise for each month
fig, ax = plt.subplots(1, 2, figsize=(14, 3))
Months = data['month'].unique()
for i in range(0, len(Months)):
    sns.countplot(data[data['month'] == Months[i]]['dayofmonth'], ax=ax[i])
    if Months[i] == 11:
        ax[i].set_title('Day Wise Flight Count for Month Nov')
    elif Months[i] == 12:
        ax[i].set_title('Day Wise Flight Count for Month Dec')
plt.show()

# plot according to day of week
plt.figure(figsize=(7, 3))
sns.countplot(data['dayofweek'])
plt.title('Day of the Week Wise Flight Count')
plt.show()
```

## Total flights with respect to date and Months for different reporting airline

```python
# Carrier-wise Flight Volume
plt.figure(figsize=(10, 5))
sns.countplot(data['reporting_airline'])
plt.title('Carrier Wise Flight Count')
plt.show()

# Carrier-wise Flight Volume for each month
plt.figure(figsize=(10, 5))
sns.countplot(data['reporting_airline'], hue=data['month'])
plt.title('Carrier Wise Flight Count for each month')
plt.show()

# Carrier-wise Flight Volume for each day of week
plt.figure(figsize=(10, 5))
sns.countplot(data['reporting_airline'], hue=data['dayofweek'])
plt.title('Carrier Wise Flight Count for each day of week')
plt.show()
```

## From the above analysis, it is found that:
1. Airline 'WH' has got the highest fligh counts among all, followed by 'AA', 'DL', and 'OO'.
2. For all Airlines, Monday stays on  top when it comes to flight count on weekdays.

## Total Flights vs Time of Day

```python
def to_hours(x):
    try:
        x = str(int(x)).zfill(4)
        return int(x[:2])
    except:
        return np.nan
data['Hour']  = data['deptime'].apply(to_hours)

def time_of_day(x):
    try:
        if x < 12:
            return "Morning"
        elif x < 17:
            return "Afternoon"
        else:
            return "Evening"
    except:
        return np.nan


data["time_of_day"] = data["Hour"].apply(time_of_day)
data["time_of_day"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Total Flights vs Time of Day")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

### From the data it is evident that most of the flights are scheduled in the morning.

## Now, let's move on to check how delay times look like for different flights.

## Average Delay of Different Airlines

```python
data.groupby('reporting_airline')['depdelayminutes'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(15, 5))
plt.title('Average delay in minutes by reporting airline')
plt.ylabel('Average delay in minutes')
plt.xlabel('Reporting airline')
plt.show()
```

### We can tell from the graph above that:
1. 'YV' has the highest average delay in flights, followed by 'G4' and 'B6'. 
2. '9E' has the lowest average delay in flight timings.

## Origin which has most delays

```python
data.groupby("origin")["depdelayminutes"].mean().sort_values(ascending=False).nlargest(10).plot(kind="bar")
plt.title("Top 10 Origin with highest average delay")
plt.ylabel("Average delay in minutes")
plt.xlabel("Origin")
plt.show()
```

#### Origin 'PPG' has  highest delay on average, 'DLG' has the minimum.

## Origins and Destinations

```python
print("Total Origin: ", data["origin"].nunique())
print("Total Destination: ", data["dest"].nunique())

# Most Popular Origin
plt.figure(figsize=(10, 5))
data["origin"].value_counts().nlargest(10).plot(kind="bar")
plt.title("Top 10 Most Popular Origin")
plt.ylabel("Flight Count")
plt.xlabel("Origin")
plt.show()

# Most Popular Destination
plt.figure(figsize=(10, 5))
data["dest"].value_counts().nlargest(10).plot(kind="bar")
plt.title("Top 10 Most Popular Destination")
plt.ylabel("Flight Count")
plt.xlabel("Destination")
plt.show()
```

### Above graphs suggests that:
 1. In terms of flight origin 'ATL' has the highest number of flights, followed by 'DFW', 'ORD', and 'DEN'.
 2. 'SEA', 'LGA', and 'LAS' have least number of flights taking off with count staying even lower than 30000 flights.
 3. Above points stay same when it comes about fligh destination.
 

# Cancellation and Diversion Analysis

```python
# Overall Cancellation and Diversion Analysis
data[['cancelled', 'diverted']].sum().plot(kind='bar', rot=0)
plt.xlabel('Cancellation/Diversion')
plt.ylabel('Count')
plt.title('Cancellation and Diversion Analysis')
plt.show()

# Cancellation and Diversion Analysis by Carrier
plt.figure(figsize=(10, 5))
data.groupby('reporting_airline')[['cancelled', 'diverted']].sum().plot(kind='bar')
plt.ylabel('Count')
plt.title('Cancellation and Diversion Analysis by Carrier')
plt.show()

# Cancellation rate by Carrier
data.groupby('reporting_airline')['cancelled'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Reporting Airline'); plt.ylabel('Cancellation Rate'); plt.title('Cancellation Rate for Each Carrier')
plt.show()

# Diversion rate by Carrier
data.groupby('reporting_airline')['diverted'].mean().sort_values(ascending=False).plot(kind='bar')
plt.xlabel('Reporting Airline'); plt.ylabel('Diversion Rate'); plt.title('Diversion Rate for Each Carrier')
plt.show()
```

### These plots give following insights:
 1. Cancellation of fligh seems to be the common options taken upon facing any anomaly.
 2. This stays the same across all airlines we have in record.
 3. Flight 'OO' has the highest number of cancellation reported.
 4. Rate to which flights usually cancell is highest for flight 'G4' and lowest for flight 'YX'. Surprisingly flight 'OO' is at 6th position.
 5. In terms of diversion rates, flight 'AS' is on top; Flight 'F9' comes last in this category.
 6. Flights 'YX', and '9E' have least rates in both cancellation and diversion.
 

## Delay Reson Analysis

```python
# Delay Analysis by Carrier
delay_cols = ['carrierdelay', 'weatherdelay', 'nasdelay', 'securitydelay', 'lateaircraftdelay']
data[delay_cols].div(data[delay_cols].sum(axis=1), axis=0).groupby(data['reporting_airline']).mean().plot(kind='bar', stacked=True)
plt.xlabel('Reporting Airline'); plt.ylabel('Proportion of Delays'); plt.title('Proportion of Delays for Each Carrier')
plt.show()

# Delay Analysis by Month
data[delay_cols].div(data[delay_cols].sum(axis=1), axis=0).groupby(data['month']).mean().plot(kind='bar', stacked=True)
plt.xlabel('Month'); plt.ylabel('Proportion of Delays'); plt.title('Proportion of Delays for Each Month')
plt.show()
```

### Points to note:
 1. 'Security' and 'Weather' dont have much imact in the porportion of delays occured.
 2. Across most of flights, delays incurred are due to Late arrival of aircraft, carrier delay, or nas delay.

## Most Popular Routes

```python
# Most Popular Route
data["route"] = data["origin"] + "-" + data["dest"]

plt.figure(figsize=(10, 5))
data["route"].value_counts().nlargest(10).plot(kind="bar")
plt.title("Top 10 Most Popular Route")
plt.ylabel("Flight Count")
plt.xlabel("Route")
plt.show()

# Most Popular Route by Carrier
plt.figure(figsize=(10, 5))
data.groupby("reporting_airline")["route"].value_counts().nlargest(10).plot(kind="bar")
plt.title("Top 10 Most Popular Route by Carrier")
plt.ylabel("Flight Count")
plt.xlabel("Route")
plt.show()

# Least Popular Route
plt.figure(figsize=(10, 5))
data["route"].value_counts().nsmallest(10).plot(kind="bar")
plt.title("Top 10 Least Popular Route")
plt.ylabel("Flight Count")
plt.xlabel("Route")
plt.show()
```

```python
# City and State Analysis
data.groupby(['origincityname', 'originstatename'])['depdelay'].mean().nlargest(10).plot(kind='barh')
plt.xlabel('Average Departure Delay'); plt.ylabel('City, State'); plt.title('Top 10 Cities by Average Departure Delay')
```

### 'Pago Pago, TT' has the highest number of delays in departure. For other states count is still comparable.

```python
sns.heatmap(data[['depdelay', 'arrdelay', 'carrierdelay', 'weatherdelay', 'nasdelay', 'securitydelay', 'lateaircraftdelay']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Delay Metrics')
plt.show()
```

### It is obvious to state that arrival of the flight is to be late when it departed late, this is supported in the graph above.
### Also, If carrier has faced delay, it can affect the timings of the departure ultimately delaying the arrival times.

```python
# Pairplot for Relationships between Delay Metrics
delay_metrics = ['DepDelay', 'ArrDelay', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
delay_metrics = [col.lower() for col in delay_metrics]  
sns.pairplot(data[delay_metrics], diag_kind='kde')
plt.suptitle('Pairplot for Relationships between Delay Metrics', y=1.02)
```

```python
# Heatmap for Departure Delays by Carrier and Month
heatmap_data = data.pivot_table(index='reporting_airline', columns='month', values='depdelay', aggfunc='mean')
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".1f", linewidths=.5)
plt.xlabel('Month'); plt.ylabel('Reporting Airline'); plt.title('Heatmap for Departure Delays by Carrier and Month')
```
